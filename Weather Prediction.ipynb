Since this dataset is an example of a Time Series data, it might be better to manually split the data than to use train_test_split

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score as R2
```

```python
def create_preds(variables,data,model):
    
    train_set = data.loc[:"2022-12-31"]
    test_set = data.loc["2023-01-01":]
    model.fit(train_set[variables],train_set["target"])
    preds = model.predict(test_set[variables])
    mae = MAE(test_set['target'],preds)
    mse = MSE(test_set['target'],preds)
    r2 = R2(test_set['target'],preds)
    combined = pd.concat([test_set['target'],pd.Series(preds,index = test_set.index)],axis = 1)
    combined.columns = ['Actual','Predictions']
    
    error = {
            "MAE" : mae,
            "MSE" : mse,
            "R2" : r2
            }
    
    return error, combined, combined.plot()
```

```python
model_1 = LinearRegression()
model_2 = Lasso()
model_3 = Ridge()
model_4 = KNN()
```

```python
variables = ["PRCP","TMAX","TMIN","TAVG"]
```

```python
create_preds(variables,data,model_1)
```

```python
create_preds(variables,data,model_2)
```

```python
create_preds(variables,data,model_3)
```

```python
create_preds(variables,data,model_4)
```

Both Linear Regression Model and Ridge Regression Model have given the best performance with a very marginal difference in accuracy between the two.

### Further improvements

```python
data
```

```python
daily_TAVG = data.expanding().mean()['TMAX']

monthly_TAVG = data.groupby(data.index.month)['TMAX'].expanding().mean().reset_index(level=0, drop=True)

# Assign the calculated averages to new columns
data['Daily_Avg'] = daily_TAVG
data['Monthly_Avg'] = monthly_TAVG
```

```python
data
```

```python
data.iloc[:45]
```

We can see that for the first month, the Daily and Monthly Average remains the same, this is because for calculating the Monthly Average, it does not have past data to look at for reference, so we can go ahead and remove these columns for further analysis.


```python
variables = ["PRCP","TMAX","TMIN","TAVG","Daily_Avg","Monthly_Avg"]
```

```python
error,combined,_ = create_preds(variables,data.iloc[30:],model_3)
```
